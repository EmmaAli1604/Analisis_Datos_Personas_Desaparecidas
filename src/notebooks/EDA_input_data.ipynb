# EDA input de data faltante a traves de distintos metodos media, mediana y moda

```python

import pandas as pd
```

```python
DATA_PROCESSED = "../../data/processed/data_processed.csv"
```

```python
def load_data():
    """Carga los datos desde el archivo CSV y devuelve un DataFrame de pandas."""
    try:
        data = pd.read_csv(DATA_PROCESSED)
        return data
    except FileNotFoundError:
        print(f"Error: El archivo {DATA_PROCESSED} no se encontró.")
        return None
    except Exception as e:
        print(f"Error al cargar los datos: {e}")
        return None
```

```python
df = load_data()
```

```python
print(df.info())
```

```python
ciegas = df[df['FECHA_DESAPARICION'].isnull() & df['FECHA_REGISTRO'].isnull()]
print(f"Casos sin ninguna fecha: {len(ciegas)}")

prop = (df['FECHA_DESAPARICION'].isnull() & df['FECHA_REGISTRO'].isnull()).mean() * 100
print(f"Proporción: {prop:.2f}%")
```

# Problema serio de calidad ed datos, se necesitan estrategias de imputación o limpieza para manejar estos casos.

```python
def limpiar_fecha(x):
    if pd.isna(x):
        return pd.NaT
    try:
        return pd.to_datetime(x, errors='coerce')
    except:
        return pd.NaT

cols_fecha = ['FECHA_NACIMIENTO', 'FECHA_DESAPARICION', 'FECHA_REGISTRO']

for col in cols_fecha:
    df[col] = df[col].apply(lambda x: limpiar_fecha(x))
```

```python
media_fecha_nacimiento = df['FECHA_NACIMIENTO'].mean()
moda_fecha_nacimiento = df['FECHA_NACIMIENTO'].mode().iloc[0]
mediana_fecha_nacimiento = df['FECHA_NACIMIENTO'].median()
```

```python
media_fecha_desaparicion = df['FECHA_DESAPARICION'].mean()
moda_fecha_desaparicion = df['FECHA_DESAPARICION'].mode().iloc[0]
mediana_fecha_desaparicion = df['FECHA_DESAPARICION'].median()
```

```python
media_fecha_registro = df['FECHA_REGISTRO'].mean()
moda_fecha_registro = df['FECHA_REGISTRO'].mode().iloc[0]
mediana_fecha_registro = df['FECHA_REGISTRO'].median()
```

```python
df_año_nacimiento = df.groupby(df['FECHA_NACIMIENTO'].dt.year)['ID_VICTIMA'].count()
df_año_nacimiento.tail(10)
```

## Fechas con media

```python

df['FECHA_NACIMIENTO'] = df['FECHA_NACIMIENTO'].fillna(media_fecha_nacimiento)
df['FECHA_DESAPARICION'] = df['FECHA_DESAPARICION'].fillna(media_fecha_desaparicion)
df['FECHA_REGISTRO'] = df['FECHA_REGISTRO'].fillna(media_fecha_registro)

# Ver el resumen estadístico
df.info()
```
